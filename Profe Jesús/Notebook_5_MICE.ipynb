```python
import seaborn as sns
import pandas as pd
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# Missing values
print(titanic.isnull().sum())
```

```python
data = titanic[['age', 'fare', 'pclass', 'sex']]
data = pd.get_dummies(data, drop_first=True)
data.head()
```

```python
imputer = IterativeImputer(max_iter=10, random_state=0)
```

```python
data_imputed = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)
```

```python
print("Datos originales")
print(data.head())
print("\nDatos imputados")
print(data_imputed.head())
```

Exercise 1:

Use MICE on the life-expectancy dataset kaggle.com/datasets/kumarajarshi/life-expectancy-who

```python
import kagglehub
DATASET_PATH = None
# Download the dataset from Kaggle
path = kagglehub.dataset_download("kumarajarshi/life-expectancy-who")

print("Path to dataset files:", path)
```

```python
life = pd.read_csv(f"{path}/Life Expectancy Data.csv")
life.head()
```

```python
# Missing values por columna
print("Missing values per column:")
print(life.isnull().sum())
```

```python

num_cols = life.select_dtypes(include=['number']).columns
cat_cols = life.select_dtypes(exclude=['number']).columns

life_dummies = pd.get_dummies(life, columns=cat_cols, drop_first=True)
life_dummies.head()
```

```python
#MICE
imputer = IterativeImputer(max_iter=10, random_state=0)
life_imputed_array = imputer.fit_transform(life_dummies)
life_imputed = pd.DataFrame(life_imputed_array, columns=life_dummies.columns)
life_imputed.head()
```

```python
#Imputados
life_imputed.head(50)
```

```python
null_counts_after = life_imputed.isnull().sum()

if null_counts_after.sum() == 0:
    print("\nNo missing values remain after MICE.")
else:
    print("\nStill missing values present.")
```

Exercise 2:

Use MICE on the Planets (seaborn) dataset to fill in missing values.

```python
# Load planets dataset
planets = sns.load_dataset("planets")
planets.head()
```

```python
# Check missing values
planets.isnull().sum()
```

```python
# Select only numeric columns
numeric_cols = planets.select_dtypes(include=['float64', 'int64']).columns
data_planets = planets[numeric_cols].copy()
data_planets.head()
```

```python
# Define MICE imputor
imputer = IterativeImputer(max_iter=10, random_state=42)
data_planets_imputed_array = imputer.fit_transform(data_planets)

data_planets_imputed = pd.DataFrame(data_planets_imputed_array, columns=data_planets.columns)
data_planets_imputed.head()
```

```python
missing_after = data_planets_imputed.isnull().sum()

if missing_after.sum() == 0:
    print("\nNo missing values remain after applying MICE to numeric columns.")
else:
    print("\nThere are still missing values present.")
```
